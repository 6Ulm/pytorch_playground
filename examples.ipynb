# PyTorch Examples

### Hand-Made Optimization

```python
import torch
import random
```

```python
dtype = torch.float
device = torch.device('cuda:0')
```

```python
n, d_in, h, d_out = 64, 1000, 100, 10
```

```python
def gauss_init(*size, grad=False):
    return torch.randn(
        *size, device=device, dtype=dtype, requires_grad=grad)
```

```python
x = gauss_init(n, d_in)
y = gauss_init(n, d_out)
```

```python
w1 = gauss_init(d_in, h, grad=True)
w2 = gauss_init(h, d_out, grad=True)
```

```python
class ReLU(torch.autograd.Function):
    
    @staticmethod
    def forward(ctx, tensor):
        ctx.save_for_backward(tensor)
        return tensor.clamp(min=0)
    
    @staticmethod
    def backward(ctx, grad_output):
        tensor, = ctx.saved_tensors
        grad_input = grad_output.clone()
        grad_input[tensor < 0] = 0
        return grad_input
```

```python
lr = 1e-6
for t in range(500):
    relu = ReLU.apply
    y_pred = relu(x.mm(w1)).mm(w2)
    loss = (y_pred - y).pow(2).sum()
    if t % 50 == 49:
        print(f'{t + 1:3d}: {loss.item():2.3f}')
    loss.backward()
    with torch.no_grad():
        w1 -= lr * w1.grad
        w2 -= lr * w2.grad
        w1.grad.zero_()
        w2.grad.zero_()
```

### High-Level Wrappers

```python
import torch.nn as nn
```

```python
n, d_in, h, d_out = 64, 1000, 100, 10
```

```python
x = torch.randn(n, d_in)
y = torch.randn(n, d_out)
```

```python
def create_model(n_in, n_out, hidden):
    return nn.Sequential(
        nn.Linear(n_in, hidden),
        nn.ReLU(),
        nn.Linear(hidden, n_out))
```

```python
model = create_model(d_in, d_out, h)
```

```python
loss_fn = nn.MSELoss(reduction='sum')
lr = 1e-4

for t in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.item())
    model.zero_grad()
    loss.backward()
    
    with torch.no_grad():
        for param in model.parameters():
            param -= lr * param.grad
```

Or even higher:

```python
model = create_model(d_in, d_out, h)
learning_rate = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

for t in range(500):
    y_pred = model(x)
    loss = loss_fn(y_pred, y)
    print(t, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

### Custom Models

```python
class DynamicNet(torch.nn.Module):
    
    def __init__(self, d_in, h, d_out):
        super().__init__()
        self.input = torch.nn.Linear(d_in, h)
        self.hid = torch.nn.Linear(h, h)
        self.out = torch.nn.Linear(h, d_out)
        
    def forward(self, x):
        h_relu = self.input(x).clamp(min=0)
        for _ in range(random.randint(0, 3)):
            h_relu = self.hid(h_relu).clamp(min=0)
        y_pred = self.out(h_relu)
        return y_pred
```

```python
n, d_in, h, d_out = 64, 1000, 100, 10
x = torch.randn(n, d_in)
y = torch.randn(n, d_out)
```

```python
model = DynamicNet(d_in, h, d_out)
```

```python
criterion = torch.nn.MSELoss(reduction='sum')
optimizer = torch.optim.SGD(model.parameters(), lr=1e-4, momentum=0.9)
for t in range(500):
    y_pred = model(x)
    loss = criterion(y_pred, y)
    print(t, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
